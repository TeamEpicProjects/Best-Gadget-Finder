```python
es = Elasticsearch([{'host': 'localhost', 'port': 9200}])
es.ping()
```

```python
# Upload csv file
with open('data/FinalProductDataset.csv', encoding='utf-8') as f:
    reader = csv.DictReader(f)
    helpers.bulk(es,reader,index='mobile')
```

```python
# Flask REST API
from flask import Flask, jsonify

app = Flask(__name__)

@app.route('/<string:Search>')
def SearchDataset(Search):
    body = {
        "_source": [
            "Name",
            "Amazon_SalesPrice",
            "Flipkart_SalesPrice",
            "Snapdeal_SalesPrice",
            "Amazon_ProductLink",
            "Flipkart_ProductLink",
            "Snapdeal_ProductLink",
            "Amazon_ImageLink",
            "Flipkart_ImageLink",
            "Snapdeal_ImageLink",
        ],
        "from": 0,
        'size': 20,
        'query': {
            "match": {
                "Name": Search
            }
        }

    }
    
    res = es.search(index='mobile', body=body)
    return jsonify(res)


app.run()
```
