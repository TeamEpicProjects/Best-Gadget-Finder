```python
import requests
from bs4 import BeautifulSoup
```

```python
phn_nm=[]
phn_pr=[]

page_num=input("Enter number of pages")
for i in range(1,int(page_num)+1):
    url="https://www.flipkart.com/search?q=phone&sid=tyy%2C4io&as=on&as-show=on&otracker=AS_QueryStore_HistoryAutoSuggest_2_5_na_na_na&otracker1=AS_QueryStore_HistoryAutoSuggest_2_5_na_na_na&as-pos=2&as-type=HISTORY&suggestionId=phone%7CMobiles&requestId=f6b7a07f-417d-4b4d-8c75-2f3041c549ff&page="+str(i)
    req=requests.get(url)
    content=BeautifulSoup(req.content,'html.parser')
    name=content.find_all('div',{"class":"_4rR01T"})
    price=content.find_all('div',{"class":"_30jeq3 _1_WHN1"})
    print("Phone in page "+str(i))
    print(len(name))  
    
    for i in name:
        phn_nm.append(i.text)
    for i in price:
        phn_pr.append(i.text)
   
    
    
```

```python
for i in phn_nm:
    print(i)
```

```python
for i in phn_pr:
    print(i)
```
