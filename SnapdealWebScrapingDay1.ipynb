## Generalize pattern

```python
def extract_record(item):
    """Extract and return data from a single record"""
    
    # URL of a product
    atag = item.find('a',{'class':'dp-widget-link hashAdded'})
    url = atag.get('href')
    
    # Image URL
    imgtag = item.find('img',{'class':'product-image'})
    imgUrl = imgtag.get('src')

    try :
        # Name of a product
        name = item.find('p',{'class':'product-title'}).text
    except AttributeError:
        return
    
    try:
        # Actual & Discount price
        actPrice = item.find('span',{'class':'lfloat product-desc-price strike'}).text
        disPrice = item.find('span',{'class':'lfloat product-price'}).text
    except AttributeError:
        actPrice = ''
        disPrice = ''
        
    result = (name,actPrice,disPrice,imgUrl,url)
    
    return result
```

```python
records = []
results = soup.find_all('div',{'class':'product-tuple-listing'})

for item in results:
    record = extract_record(item)
    if record:
        records.append(record)
```

## Extract Star rating
using product link

```python
#Product URL
rate_url = []

for i in range(0,20):
    rate_url.append(records[i][0])
    
print(rate_url)
```

```python
def get_star_rating(url):
    """Extract and return product name & rating"""

    driver.get(f'{url}')
    rating_result = BeautifulSoup(driver.page_source, 'html.parser')

    try:
        # Product Name
        product_name = rating_result.find('h1', {'itemprop': 'name'}).text.strip()
    except AttributeError:
        return

    try:
        # Product Rating
        rating_count = rating_result.find('span', {'class': 'avrg-rating'}).text.strip()
    except AttributeError:
        rating_count = '0'

    values = (product_name, rating_count)

    return values
```

```python
# Store Name & Rating
df = []

# Iterating the URL List
for item in rate_url:
    record = get_star_rating(item)
    
    if record:
        df.append(record)
        
        # sleep for 1sec
        time.sleep(1)
```

```python
df = list(set(df))
df[0]
```

```python
Product_names = []
Product_rating = []

for i in range(0,20):
    prod_name2 = df[i][0]
    pro_rat = df[i][1]
    
    Product_names.append(prod_name2)
    Product_rating.append(pro_rat)
    
Product_names = pd.DataFrame(Product_names, columns = ['Product_name'])
Product_rating = pd.DataFrame(Product_rating, columns = ['Rating'])
new_data = pd.concat([Product_names,Product_rating],axis=1)
```
